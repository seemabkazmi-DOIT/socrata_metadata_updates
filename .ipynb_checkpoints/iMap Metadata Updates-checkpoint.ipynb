```python
import requests
from requests.auth import HTTPBasicAuth
import json
from secret_data import userpass 
import pandas as pd 
import re 
# credentials
domain = "opendata.maryland.gov"
dataset_id = "ynzp-27um"
username = userpass[0]
password = userpass[1]

# Metadata endpoint URL
metadata_url = f"https://{domain}/api/views/{dataset_id}"

# Make the request
response = requests.get(metadata_url, auth=HTTPBasicAuth(username, password))

# Check the response
if response.status_code == 200:
    metadata = response.json()
    print("Metadata retrieved successfully!")
    print(metadata)
else:
    print(f"Failed to retrieve metadata. Status code: {response.status_code}")
    print(response.text)
```

```python
url = metadata['attributionLink']
base_url = url.rsplit('/', 1)[0]
base_number = url.rsplit('MapServer', 1)[1]
imap = pd.read_csv('iMAP4-ServiceInventory-Prod.csv')
odp = pd.read_csv('Socrata Imap Link Replacements.csv')

new_url_number = "https://mdgeodata.md.gov/imap/rest/services/Hydrology/MD_ShorelineInventory/MapServer/0"
metadata['attributionLink'] = new_url_number 

# New URL to insert
new_url = "https://mdgeodata.md.gov/imap/rest/services/Hydrology/MD_ShorelineInventory/MapServer"

# Replace the URL following "Map Service Link:"
metadata['description'] = re.sub(
    r"(Map Service Link:\s*)(http[s]?://\S+)",
    rf"\1{new_url}",
    metadata['description']
)

print(base_number)
```

```python
update_response = requests.put(
    metadata_url,
    auth=HTTPBasicAuth(username, password),
    headers={"Content-Type": "application/json"},
    data=json.dumps(metadata)
)

if update_response.status_code in [200, 202]:
    print("Source link updated successfully!")
else:
    raise Exception(f"Failed to update metadata: {update_response.status_code}, {update_response.text}")
```

```python
import requests
import json
import re
from requests.auth import HTTPBasicAuth

for i in odp['U ID']:
    metadata_url = f"https://{domain}/api/views/{i}.json"
    response = requests.get(metadata_url, auth=HTTPBasicAuth(username, password))

    if response.status_code == 200:
        metadata = response.json()
        print(f"Metadata retrieved for {i}")
    else:
        print(f"Failed to retrieve metadata for {i}. Status code: {response.status_code}")
        print(response.text)
        continue  # skip to next uid

    # Get old URL
    url = metadata.get('attributionLink', '')
    if not url or 'MapServer' not in url:
        print(f"No valid MapService link found for {i}")
        continue

    try:
        base_url = url.rsplit('/MapServer', 1)[0]
        matching_rows = imap[imap['iMAP 3.0'] == base_url]
        if not matching_rows.empty:
            new_base_url = matching_rows['iMAP 4.0'].iloc[0]
            new_url = new_base_url + "/MapServer"
        else:
            print(f"No match found in iMAP table for {base_url}")
            continue
    except Exception as e:
        print(f"Error processing iMAP mapping: {e}")
        continue

    # Update metadata
    metadata['attributionLink'] = new_url
    metadata['description'] = re.sub(
        r"(Map Service Link:\s*)(http[s]?://\S+)",
        rf"\1{new_url}",
        metadata.get('description', '')
    )

    update_response = requests.put(
        metadata_url,
        auth=HTTPBasicAuth(username, password),
        headers={"Content-Type": "application/json"},
        data=json.dumps(metadata)
    )

    if update_response.status_code in [200, 202]:
        print(f"✅ Source link updated for {i}")
    else:
        print(f"❌ Failed to update {i}: {update_response.status_code}, {update_response.text}")
```
